# **DATA ANALYSIS**

```python
climate_data['DATE'] = pd.to_datetime(climate_data['DATE'])
```

```python
temperature_columns = ['DATE', 'DailyAverageDewPointTemperature',
                       'DailyAverageDryBulbTemperature', 'DailyAverageWetBulbTemperature',
                       'DailyMaximumDryBulbTemperature', 'DailyMinimumDryBulbTemperature']
```

```python
temperature_data = climate_data[temperature_columns]
```

```python
plt.figure(figsize=(10, 6))

plt.plot(temperature_data['DATE'], temperature_data['DailyAverageDryBulbTemperature'], label='Daily Avg Dry Bulb Temp', color='red')
plt.plot(temperature_data['DATE'], temperature_data['DailyMaximumDryBulbTemperature'], label='Daily Max Dry Bulb Temp', color='blue')
plt.plot(temperature_data['DATE'], temperature_data['DailyMinimumDryBulbTemperature'], label='Daily Min Dry Bulb Temp', color='green')

plt.title('Temperature Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
temperature_data['Year'] = temperature_data['DATE'].dt.year
temperature_data['Month'] = temperature_data['DATE'].dt.month
```

```python
monthly_avg = temperature_data.groupby(['Year', 'Month']).mean()
yearly_avg = temperature_data.groupby('Year').mean()
```

```python
monthly_avg['Temperature Anomaly'] = monthly_avg['DailyAverageDryBulbTemperature'] - yearly_avg['DailyAverageDryBulbTemperature'].mean()
```

```python
correlation_matrix = temperature_data.corr()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(monthly_avg.index.get_level_values('Year'), monthly_avg['DailyAverageDryBulbTemperature'], label='Monthly Avg Dry Bulb Temp', color='orange')
plt.title('Monthly Average Temperature Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(yearly_avg.index, yearly_avg['DailyAverageDryBulbTemperature'], label='Yearly Avg Dry Bulb Temp', color='purple')
plt.title('Yearly Average Temperature Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Correlation Matrix:")
print(correlation_matrix)
```
